# Aircraft Speed Analysis. Most planes are travelling at the mean speed of 6.57122 miles/min

No of NA inside each columns of the dataframe

```python
speed_df=raw_data
speed_df.isnull().sum()
```

create another column call speed. Taking the speed divide by the airtime for mile/min

```python
speed_df['speed']=speed_df.distance/speed_df.air_time
speed_df
```

```python
speed_df['speed'].describe()
```

drop all rows that contain NAN under speed column

```python
speed_df=speed_df.dropna(subset=['speed'])
speed_df
```

create a new dataframe with only four columns by dropping all the rest of the columns. The table show the averge speed of the carriers flying from the 3 airports to each destinations.

```python
#yy.groupby('month').size().describe()
yy=speed_df
yy=yy.drop(['year','month','day','dep_time','sched_dep_time','dep_delay','arr_time','sched_arr_time','arr_delay','flight','tailnum','air_time','distance','hour','minute','time_hour'],axis=1)
yy.groupby(['origin','carrier','dest']).mean().unstack().reset_index()
```
